```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

Check if there is null values

```python
df.isnull().sum()
```

```python
df.shape
```

as we see there is 177 raw of Age cloumn is null and number of raw data 891 so we can change null data by mean and there 2 null values in Embarked so we we can delete just this 2 raws and there is 687 null values in Cabin and i think we dont interset about this column so we will drop it

```python
df.drop('Cabin',inplace=True,axis = 1)
```

```python
df.head(22)
```

```python
# df.Age.fillna(df.Age.mean())
# import the modules we need
from sklearn.preprocessing import Imputer
# make a new imputer 
imputer = Imputer(missing_values='NaN', strategy='mean',axis=0)
df[['Age']] = imputer.fit_transform(df[['Age']])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna(axis = 0, how ='any')
```

```python
df.shape
```

now i see there coloumns we dont need so we will drop it

```python
df = df.drop(['Name','PassengerId','Ticket','Fare','Embarked'],axis=1 )
df
```

now we see all columns digital data but sex is string so we aill make encoder for this coulmn

```python
from sklearn.preprocessing import LabelEncoder ,OneHotEncoder
encoder = LabelEncoder()
df.Sex = encoder.fit_transform(df['Sex'])
df
```

and there is just 2 Sex category so we dont need to use onehotencoder

now we need to split the data to features and goals and after that to train and test

```python
# features 
x = df.drop('Survived' , axis=1)
x
```

```python
y = df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.2, random_state = 0)
print(len(x_train))
print(len(x_test))
```

now test prediction

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
goal_predict = model.predict(x_test)
goal_predict
```

```python
plt.plot(x_test,'.', goal_predict, '-')
plt.title('our first simple model')
plt.xlabel('x')
plt.ylabel('Serv')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error ,r2_score
print(mean_squared_error(y_train, model.predict(x_train)))
print(mean_squared_error(y_test, model.predict(x_test)))
print(model.score(x_train, y_train))
print(model.score(x_test,y_test))
y_pred = model.predict(x_test)
print(model.score(x_test,y_pred))
print(r2_score(y_test,y_pred))
# print(r2_score(x_test,y_test))
```
